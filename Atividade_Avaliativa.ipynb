#Questão 1- Classificação de Temperaturas (Condicionais + Loop)

Dada a lista de temperaturas registradas durante a semana:

temperaturas = [18, 22, 30, 35, 28, 15, 40]

Classifique cada dia como:

* Abaixo de 20 → "Frio"

* Entre 20 e 30 → "Agradável"

* Acima de 30 → "Quente"

Tarefas:
1. Criar uma lista com as classificações
2. Contar quantos dias houve em cada categoria


```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]

classificacoes = []

for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
    elif 20 <= temp <= 30:
        classificacoes.append("Agradável")
    else:
        classificacoes.append("Quente")

print(f"Classificações por dia: {classificacoes}")

# Contar quantos dias houve em cada categoria
contagem_frio = classificacoes.count("Frio")
contagem_agradavel = classificacoes.count("Agradável")
contagem_quente = classificacoes.count("Quente")

print(f"\nNúmero de dias 'Frio': {contagem_frio}")
print(f"Número de dias 'Agradável': {contagem_agradavel}")
print(f"Número de dias 'Quente': {contagem_quente}")
```

#Questão 2 — Sistema de Avaliação de Alunos

Dada a lista de notas finais:

notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

Classifique:

* < 5 → "Reprovado"
* 5 a 7 → "Recuperação"
* 7 → "Aprovado"

Tarefas:
1. Gerar lista de classificações
2. Calcular:
o Quantidade de aprovados
o Percentual de reprovados

```python
notas = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]

classificacoes_notas = []

for nota in notas:
    if nota < 5:
        classificacoes_notas.append("Reprovado")
    elif 5 <= nota < 7:
        classificacoes_notas.append("Recuperação")
    else: # nota >= 7
        classificacoes_notas.append("Aprovado")

print(f"Classificações das notas: {classificacoes_notas}")

# Calcular quantidade de aprovados
quantidade_aprovados = classificacoes_notas.count("Aprovado")
print(f"Quantidade de aprovados: {quantidade_aprovados}")

# Calcular percentual de reprovados
quantidade_reprovados = classificacoes_notas.count("Reprovado")
total_alunos = len(notas)
percentual_reprovados = (quantidade_reprovados / total_alunos) * 100 if total_alunos > 0 else 0
print(f"Percentual de reprovados: {percentual_reprovados:.2f}%")
```

# Questão 3 — Monitoramento de Consumo de Energia

Consumo diário (kWh):

consumo = [120, 95, 200, 180, 75, 220, 160]

Regras:
* < 100 → "Baixo"
* 100–180 → "Moderado"
* 180 → "Alto"

Tarefas:
1. Filtrar dias com consumo alto
2. Calcular:
o Total de consumo da semana
o Média semanal
3. Exibir alerta se houver mais de 2 dias com consumo alto

```python
consumo = [120, 95, 200, 180, 75, 220, 160]

# 1. Filtrar dias com consumo alto
dias_consumo_alto = []
for c in consumo:
    if c > 180: # Interpreting '180 -> Alto' as > 180
        dias_consumo_alto.append(c)

print(f"Dias com consumo alto (kWh): {dias_consumo_alto}")

# 2. Calcular total de consumo da semana
total_consumo = sum(consumo)
print(f"Total de consumo da semana (kWh): {total_consumo}")

# 2. Calcular média semanal
media_semanal = total_consumo / len(consumo) if len(consumo) > 0 else 0
print(f"Média semanal de consumo (kWh): {media_semanal:.2f}")

# 3. Exibir alerta se houver mais de 2 dias com consumo alto
if len(dias_consumo_alto) > 2:
    print(f"ALERTA: Há {len(dias_consumo_alto)} dias com consumo alto, o que excede o limite de 2 dias.")
else:
    print(f"Não há alerta de consumo alto (apenas {len(dias_consumo_alto)} dia(s) com consumo alto).")
```

# Questão 4 — Simulação de Carrinho de Compras

4- Lista de preços:

preços = [50, 120, 30, 200, 80, 15]

Regras de desconto:

* Item < 50 → 5% desconto
* 50–150 → 10% desconto
* 150 → 15% desconto

Tarefas:
1. Criar lista com preços finais

2. Calcular:
* Total da compra com desconto
* Economia total obtida


```python
precos = [50, 120, 30, 200, 80, 15]

precos_finais = []
total_original = sum(precos)
total_com_desconto = 0

for preco in precos:
    desconto = 0
    if preco <= 50: # Ajustado para incluir 50 com 5% de desconto
        desconto = 0.05
    elif preco < 150: # Ajustado para 10% de desconto entre 50 (exclusivo) e 150 (exclusivo)
        desconto = 0.10
    else: # 15% de desconto para 150 ou mais
        desconto = 0.15

    preco_final = preco * (1 - desconto)
    precos_finais.append(preco_final)
    total_com_desconto += preco_final

print(f"Lista com preços finais: {precos_finais}")

# Calcular total da compra com desconto
print(f"Total da compra com desconto: {total_com_desconto:.2f}")

# Calcular economia total obtida
economia_total = total_original - total_com_desconto
print(f"Economia total obtida: {economia_total:.2f}")
```
